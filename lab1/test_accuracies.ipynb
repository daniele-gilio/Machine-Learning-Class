```python
import matplotlib.pyplot as plt
import numpy as np

n, bayes, svm, logreg=np.loadtxt("test_acc.txt", unpack=True)

plt.plot(n, bayes, label="Bayes", marker=".")
plt.plot(n, svm, label="SVM", marker=".")
plt.plot(n,logreg, label="LogReg", marker=".")
plt.xlabel("Dictionary Size")
plt.ylabel("Test Accuracy")
plt.xticks(n)
plt.grid()
plt.legend()
plt.savefig("test_acc.png")
#plt.show()
```
